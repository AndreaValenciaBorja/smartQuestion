**ANALISIS BIVARIADO**

El data set cuenta con la informacion de los paciente que ingresaron a la Unidad de cuidado intensivo adultos en el primer semestre del 2024

```python
# Importar librerías
import pandas as pd
import altair as alt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Se carga el dataset
df = pd.read_excel('/content/pertinencia_uci.xlsx')
```

```python
# Mostrar las primeras 5 filas del dataset utilizando el método dataframe.head()
df.head(6)
```

```python
# Comprobar el tipo de datos
df.dtypes
```

```python
# Obtener el resumen estadístico de la columna 'Edad' de los pacientes que ingresaron a UCI los ultimos 6 meses
resumen = df['Edad'].describe()

# Mostrar el resumen
print(resumen)
```

```python
# Crear el boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Edad'], color='lightblue')
plt.title('Boxplot de Edad')
plt.xlabel('Edad')

plt.show()
```

La edad promedio se sitúa en 56.81 años, La desviación estándar de 18.76 años sugiere que existe una variabilidad considerable en las edades, con algunos individuos mucho más jóvenes o mayores que el promedio.Y la distribución de las edades podría ser aproximadamente simétrica, con una ligera tendencia a la derecha (positivo sesgo), ya que la diferencia entre el tercer cuartil (75%) y la mediana es ligeramente mayor que la diferencia entre la mediana y el primer cuartil (25%).

```python
# Obtener el resumen estadístico del criterio general de ingreso a UCI de los
# pacientes que estuvieron en la unidad los ultimos 6 meses
df['Criterio general'].value_counts()
```

```python
# Se grafica la distribucion y comportamiento de los diferentes criterios
# generales usados al clasificar el paciente en su ingreso a UCI
class_counts = df['Criterio general'].value_counts()
class_counts.plot(kind='bar')
plt.title('Distribución del Criterio general de ingreso a UCI')
plt.xlabel('Tipos de Criterio general')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Calcular estadísticas descriptivas agrupadas por la variable categórica
# Aquí, agrupamos por 'Criterio general' y calculamos la media de 'Edad' para cada categoría
estadisticas = df.groupby('Criterio general')['Edad'].describe()
print(estadisticas)
```

```python
# 2. Visualizar la comparación con un boxplot
# El boxplot muestra la distribución de 'Edad' para cada categoría en 'Criterio general'
plt.figure(figsize=(8, 6))
sns.boxplot(y='Criterio general', x='Edad', data=df)
plt.title('Comparación de la Edad por Criterio de UCI')
plt.ylabel('Criterio general')  # Etiqueta para el eje y (categorías)
plt.xlabel('Edad')              # Etiqueta para el eje x (variable numérica)
plt.show()
```

# **ANALISIS BIVARIADO**

El grupo **CARDIOVASCULAR** tiene la mayor cantidad de observaciones con una media de edad de 61.18 años y una desviación estándar de 16.48 años, indicando edades avanzadas y una distribución amplia. El criterio **GASTROINTESTINAL**, aunque con menos observaciones, presenta una mayor dispersión con una desviación estándar de 21.87 años. El criterio **ENDOCRINO** tiene una media de 54.81 años y una desviación estándar de 18.59 años, menos dispersa que el criterio **RENAL**, que muestra una alta variabilidad con una desviación estándar de 27.65 años. **GRAN QUEMADOS** tiene una edad media de 50.56 años y una baja desviación estándar de 5.75 años, indicando homogeneidad. El criterio de **CHOQUE** se considera como dato atipico dado que no se cuenta con suficientes casos en esta categoria. EL criterio **RESPIRATORIO** y **QUIRURGICO** presentan medias de edad de 56.28 y 53.93 años, respectivamente, con una dispersión moderada.
